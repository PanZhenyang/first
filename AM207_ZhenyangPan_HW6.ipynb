### Part A

First we need to proof: the product of Poisson and Gamma is Gamma      
For y    

$$Gamma(a, b) * \text{Poisson}(\lambda) = \frac{b^a}{\Gamma(a)}\lambda^{a-1}e^{-b\lambda}\frac{\lambda^y}{y!}e^{-\lambda} \propto \lambda^{a-1}e^{-b\lambda}\lambda^ye^{-\lambda} \propto \lambda^{a+y-1}e^{-(b+1)\lambda} \propto \text{Gamma}(a+y, b+1)$$

We could see that Gamma is the conjugate prior of Poisson

And for   
$$\theta \sim Gamma(a, b)$$
The posterior distribution is:  
$$\theta|y \sim Gamma(a+\sum^{N}_{j=1}y_j, b+n)$$

### Part B

The posterior distribution
$$ P(\theta, \alpha, \beta|Y) \propto P(\alpha)P(\beta)\prod_{i=1}^NP(y_i|\theta, \alpha, \beta)\prod_{i=1}^NP(\theta_i|\alpha, \beta)
\propto P(\alpha)P(\beta)\prod_{i=1}^{N}\left\{\frac{\beta^\alpha}{\Gamma(\alpha)}\theta_i^{\alpha-1}e^{-\beta\theta_i}\frac{(n_i\theta_i)^{y_i}}{y_i!}e^{-n_i\theta_i}\right\}$$

Treating alpha and beta in the posterior distribution as constants:

$$p(\lambda|\alpha, \beta, y_j, n_j)=\prod^{N}_{j=1}(\theta_j n_j)^{y_j+\alpha-1}e^{-(\theta_j n_j)(\beta+n)}$$

Which follows a Gamma distribution, 

$$p( \lambda_j | \alpha, \beta, y_j, n_j)\propto Gamma(\alpha + y_j, \beta + n_j)$$

By removing all terms not having a relationship with alpha:

$$p(\alpha|\lambda, \beta, y_j, n_j)=\prod^{N}_{j=1}(\theta_j n_j)^{\alpha-1}\alpha^{a_{\alpha}-1}e^{\alpha b_{\alpha}}$$

By removing all terms not having a relationship with beta:

$$p(\beta|\lambda, a, y_j, n_j) \propto \frac{{b_\beta}^{a_\beta}}{\Gamma(a_\beta)}\beta^{a_\beta-1}e^{-b_\beta\beta}\beta^{N\alpha}\prod_{i=1}^{N}e^{-\beta\theta_i} \propto \text{Gamma}(a_\beta+N\alpha, b_\beta+\sum_{j=1}^N\theta_j)$$

MH algorithm based on part a,b and c

```python
burnin=10000
n=100000
N=burnin+n
#length of all data in NY
l = 62

alp=np.empty(N)
beta=np.empty(N)
lam=np.empty((N,l))

#initilize
alp[0]=441
beta[0]=100000
lam[0] = NY['dc'].values/NY['pop'].values
#count for accept 
acc_apl = 0

for i in np.arange(1,N):     
    #select alpha
    star_ = norm.rvs(alp[i-1], step)
    num_ = (star_-1)*np.sum(np.log(lam[i-1]))+(a_alpha-1)*np.log(star_)-(star_*b_alpha)
    den_ = (alp[i-1]-1)*np.sum(np.log(lam[i-1]))+(a_alpha-1)*np.log(star_)-(alp[i-1]*b_alpha)
    U=np.log(np.random.rand())
    
    if U <= num_ - den_ and star_ >0:
        acc_apl+=1
        alp[i] = star_
    else:
        alp[i] = alp[i-1]
    
    beta[i] = scipy.stats.gamma.rvs(a_beta,1.0/(np.sum(lam[i-1,:]) + b_beta))
    lam[i,:] = scipy.stats.gamma.rvs(alp[i] + NY['dc'].values, 1.0/(beta[i] + NY['pop'].values))
```
